### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import roc_curve, auc

# Load data and split into features and target variable
X = df_imputed_KNN_scaled[['gender', 'age', 'heart_rate_min', 'heart_rate_mean', 'mbp_min',
       'mbp_mean', 'sbp_min', 'dbp_min', 'dbp_max', 'dbp_mean',
       'temperature_min', 'temperature_mean', 'lactate_min', 'ph_min',
       'calcium_min', 'sodium_max', 'glucose_min', 'wbc_max',
       'total_protein_max', 'pt_min', 'alp_max', 'alp_min', 'ast_max',
       'bilirubin_direct_max', 'bilirubin_indirect_max',
       'bilirubin_indirect_min', 'urineoutput', 'sofa_coagulation',
       'sofa_cardiovascular', 'sofa_renal', 'charlson_comorbidity_index']]
y = df[['outcome']]

# Split the data into a training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Fit the logistic regression model with cross-validation
params = {'C': [0.01, 0.1, 1, 10, 100]}
logreg = LogisticRegression(max_iter = 10000)
grid_search = GridSearchCV(logreg, params, cv=5)
grid_search.fit(X_train_resampled, y_train_resampled)
scores = cross_val_score(logreg, X_train_resampled, y_train_resampled, cv=5)

# Print the cross-validation scores
print('Cross-validation scores:', scores)
print('Mean cross-validation score:', scores.mean())

# Predict the target variable on the test data
y_pred = grid_search.predict(X_test)

# Evaluate the performance of the model on the test data
print('Classification report:\n', classification_report(y_test, y_pred))
print('Confusion matrix:\n', confusion_matrix(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, grid_search.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print("AUC:", roc_auc)
```

```python
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Oversample using SMOTE
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Set up the decision tree classifier
dtc = DecisionTreeClassifier()

# Define the hyperparameters to tune
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2,3,4,5,6],
    'min_samples_leaf': [1, 2, 3, 4, 5]
}

# Use grid search with cross-validation to find the best hyperparameters
grid_search = GridSearchCV(dtc, params, cv=5)
grid_search.fit(X_train_resampled, y_train_resampled)

# Evaluate the performance of the model using classification report and confusion matrix
y_pred = grid_search.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

# Compute the ROC curve and AUC score
y_pred_prob = grid_search.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = roc_auc_score(y_test, y_pred_prob)

# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
print(grid_search.best_estimator_.get_params())
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Oversample minority class using SMOTE
sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

# Define random forest model
rfc = RandomForestClassifier()

# Define parameter grid for grid search
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(rfc, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Print best parameters and best score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

# Predict target variable for test set
y_pred = grid_search.predict(X_test)

# Print classification report and confusion matrix
print("Classification report:\n", classification_report(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))

# Calculate false positive rate and true positive rate for ROC curve
fpr, tpr, _ = roc_curve(y_test, y_pred)

# Calculate AUC for ROC curve
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

### SVM

```python
from sklearn.model_selection import cross_val_predict
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Oversampling the minority class using SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Creating an SVM model with radial basis function kernel
svm = SVC(kernel='rbf', probability=True)
svm.fit(X_train_resampled, y_train_resampled)
```

```python
# Predicting on the test set
y_test_pred = svm.predict(X_test)

# Printing the classification report and confusion matrix for the test set
print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))

# Plotting the ROC curve for the test set
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, svm.decision_function(X_test))
roc_auc_test = roc_auc_score(y_test, svm.decision_function(X_test))

plt.plot(fpr_test, tpr_test, label='Test ROC curve (area = %0.2f)' % roc_auc_test)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

### Ada Boost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import StratifiedKFold,cross_val_predict
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Oversample the minority class using SMOTE
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# Perform hyperparameter tuning using GridSearchCV
param_grid = {'n_estimators': [50, 100, 200],
              'learning_rate': [0.01, 0.1, 1]}
ada = AdaBoostClassifier()
grid_search = GridSearchCV(ada, param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
best_estimator = grid_search.best_estimator_

# Evaluate the model on the test set
y_pred = best_estimator.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

# Plot the ROC curve
y_pred_prob = best_estimator.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = roc_auc_score(y_test, y_pred_prob)

plt.figure(figsize=(8,8))
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

### Gradient Boost

```python
# import necessary libraries
from sklearn.ensemble import GradientBoostingClassifier

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# perform oversampling using SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# define the model and parameters to be tuned using GridSearchCV
gb = GradientBoostingClassifier(random_state=42)
params = {'n_estimators': [50, 100, 150], 'learning_rate': [0.01, 0.1, 1]}

# perform GridSearchCV with 5-fold cross validation
grid_search = GridSearchCV(gb, params, cv=5, n_jobs=-1)
grid_search.fit(X_train_resampled, y_train_resampled)

# print the best parameters
print("Best Parameters: ", grid_search.best_params_)

# fit the model with the best parameters
gb = GradientBoostingClassifier(random_state=42, **grid_search.best_params_)
gb.fit(X_train_resampled, y_train_resampled)

# make predictions
y_pred = gb.predict(X_test)

# generate classification report and confusion matrix
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))

# calculate the predicted probabilities and generate ROC curve
y_prob = gb.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

# plot the ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
